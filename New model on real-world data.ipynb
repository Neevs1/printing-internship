```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pakka_data = pd.read_excel('real job dataset ( pakka wala ).xlsx')
pakka_data
```

```python
print(pakka_data.isnull().sum())
```

```python
pakka_data = pakka_data.dropna()
print(pakka_data.isnull().sum())
```

```python
pakka_data.dtypes
```

```python
fivemm_data = pd.read_excel('Sample data for 5mm (34_).xlsx')
```

```python
fivemm_data
```

```python
pakka_data['Ink key zero setting'] = pakka_data['Ink key zero setting'].str.replace("mm","",regex=False).astype(int)
fivemm_data['Ink key zero setting'] = fivemm_data['Ink key zero setting'].str.replace("mm","",regex=False).astype(int)
pakka_data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
fivemm_data.rename(columns={'Paper Type': 'Paper type','Delta E after ': 'Delta E after'}, inplace=True)
fivemm_data
```

```python
pakka_data = pd.concat([pakka_data,fivemm_data])
```

```python
pakka_data
```

```python
pakka_data['Paper type']=pakka_data['Paper type'].map({'Coated':0,'Uncoated':1})
pakka_data['Color']=pakka_data['Color'].map({'Cyan':0,'Magenta':1,'Yellow':2,'Black':3})
pakka_data
```

```python
pakka_data['Job Number'] = pakka_data['Job Number'].astype(str).str.split(' ').str[0].astype(int)

pakka_data
```

```python
pakka_data.cov()
```

```python
x=pakka_data.drop('final ink key setting',axis=1)
y=pakka_data['final ink key setting']
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression, RidgeCV
```

```python
linear = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=41)
```

```python
linear.fit(x_train,y_train)
```

```python
linear.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_pred = linear.predict(x_test)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
rf = RandomForestRegressor().fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python
gb = GradientBoostingRegressor().fit(x_train,y_train)
gb.score(x_train,y_train)
```

```python
y_pred = gb.predict(x_test)
```

```python
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python

scores = []
for i in range(1,10):
    gb = GradientBoostingRegressor(learning_rate=(i*0.1)).fit(x_train,y_train)
    scores.append(gb.score(x_test,y_test))
plt.plot(range(1, 10), scores)
plt.xlabel('Value of Learning Rate for XGB')
plt.ylabel('Testing Accuracy')
```

```python
#Changing n_predictors for an optimal learning rate
scores = []
for i in range(1,200):
    gb = GradientBoostingRegressor(learning_rate=0.4,n_estimators=i).fit(x_train,y_train)
    scores.append(gb.score(x_test,y_test))
plt.plot(range(1, 200), scores)
plt.xlabel('Value of Estimators for XGBoost')
plt.ylabel('Testing Accuracy')
```

```python
print('Max score is',max(scores))
print('Required number of estimators',scores.index(max(scores))+1)
```

```python
y_pred = gb.predict(x_test)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python
ridge = RidgeCV(alphas=np.logspace(-3, 3, 7)).fit(x_train,y_train)
```

```python
ridge.score(x_train,y_train)
```

```python
y_pred = ridge.predict(x_test)
```

```python
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python
estimators = [
    ('lr',LinearRegression()),
    ('rf', RandomForestRegressor(n_estimators=170, random_state=42)),
    ('gbr', GradientBoostingRegressor(n_estimators=199, learning_rate=0.1, random_state=42)),
    ('ridge', RidgeCV(alphas=np.logspace(-3, 3, 7)))
]


stack = StackingRegressor(
    estimators=estimators,
    final_estimator=RidgeCV(),
    cv=5,
    passthrough=True
)
stack.fit(x_train,y_train)
```

```python
y_pred = stack.predict(x_test)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python
param_grid = {
    
    'rf__n_estimators': [50, 200],
    'rf__max_depth': [None, 10],
    
    
    'gbr__learning_rate': [0.01, 0.1],
    'gbr__n_estimators': [100, 200],
    
    'final_estimator__alphas': [0.1, 1.0, 10.0]
}
```

```python
grid = GridSearchCV(
    estimator=stack, 
    param_grid=param_grid, 
    cv=5, 
    scoring='neg_mean_squared_error',
    n_jobs=-1 
)

grid.fit(x_train, y_train)

print(f"Best Params: {grid.best_params_}")
```

```python
y_pred = grid.predict(x_test)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```
