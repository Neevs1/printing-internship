```python
import pandas as pd
import numpy as np
```

```python
pakka_data = pd.read_excel('real job dataset ( pakka wala ).xlsx')
pakka_data
```

```python
print(pakka_data.isnull().sum())
```

```python
pakka_data = pakka_data.dropna()
print(pakka_data.isnull().sum())
```

```python
pakka_data.dtypes
```

```python
fivemm_data = pd.read_excel('Sample data for 5mm (34_).xlsx')
```

```python
fivemm_data
```

```python
pakka_data['Ink key zero setting'] = pakka_data['Ink key zero setting'].str.replace("mm","",regex=False).astype(int)
fivemm_data['Ink key zero setting'] = fivemm_data['Ink key zero setting'].str.replace("mm","",regex=False).astype(int)
pakka_data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
fivemm_data.rename(columns={'Paper Type': 'Paper type','Delta E after ': 'Delta E after'}, inplace=True)
fivemm_data
```

```python
pakka_data = pd.concat([pakka_data,fivemm_data])
```

```python
pakka_data
```

```python
pakka_data['Paper type']=pakka_data['Paper type'].map({'Coated':0,'Uncoated':1})
pakka_data['Color']=pakka_data['Color'].map({'Cyan':0,'Magenta':1,'Yellow':2,'Black':3})
pakka_data
```

```python
pakka_data['Job Number'] = pakka_data['Job Number'].astype(str).str.split(' ').str[0].astype(int)

pakka_data
```

```python
pakka_data.cov()
```

```python
x=pakka_data.drop('final ink key setting',axis=1)
y=pakka_data['final ink key setting']
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=41)
```

```python
linear.fit(x_train,y_train)
```

```python
linear.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_pred = linear.predict(x_test)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
rf = RandomForestRegressor().fit(x_train,y_train)
rf.score(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```

```python
gb = GradientBoostingRegressor().fit(x_train,y_train)
gb.score(x_train,y_train)
```

```python
y_pred = gb.predict(x_test)
```

```python
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred)

print("\n--- Test Set Accuracy ---")
print(f"MSE: {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"R^2 Score: {test_r2:.4f}")
```
